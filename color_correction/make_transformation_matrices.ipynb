```python
from plantcv import plantcv as pcv
import cv2
import numpy as np
import os
pcv.params.debug = "plot"
```

# Target Image

```python
target_img, target_path, target_filename = pcv.readimage(filename="TM009_B_092115_VIS_SV_0_z1500_h2_g0_e100_281446_0.png",
                                                        debug="plot")
target_mask = cv2.imread("color_card_mask_z1500.png", -1) # mask must be read in "as-is" include -1
```

# Source: TM009 z1500

```python
source_img, source_path, source_filename = pcv.readimage(filename="TM009_B_092115_VIS_SV_0_z1500_h2_g0_e100_258104_0.png",
                                                        debug="plot")
source_mask = cv2.imread("color_card_mask_z1500.png", -1) # mask must be read in "as-is" include -1
```

```python
#.npz files containing target_matrix, source_matrix, and transformation_matrix will be saved to the 
# output_directory file path
output_directory = "tm009_z1500_transformation_matrices"
if not os.path.exists(output_directory):
    os.mkdir(output_directory)
```

```python
target_matrix, source_matrix, transformation_matrix, corrected_img = pcv.transform.correct_color(target_img, target_mask, source_img, source_mask, output_directory)
```
